```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "order_id":[101,102,103,104,105,106,107,108,109,110],
    "customer":["Alice","Bob","Alice","David","Emma","Bob","Alice","Emma","David","Bob"],
    "city":["Amsterdam","Rotterdam","Amsterdam","Utrecht","Eindhoven",
            "Rotterdam","Amsterdam","Eindhoven","Utrecht","Rotterdam"],
    "category":["Fruit","Snacks","Fruit","Drinks","Fruit",
                "Drinks","Snacks","Fruit","Drinks","Snacks"],
    "price":[3.5,5.0,2.0,4.0,6.5,3.0,2.5,np.nan,7.0,4.5],
    "quantity":[5,2,8,1,3,4,6,2,np.nan,5],
    "delivery_time":[15,25,10,30,18,20,12,22,35,16]
})

df["total_price"] = df["price"] * df["quantity"]
```

```python
# Problem 1
# Show:
# - dataframe shape
# - column names
# - summary statistics

df.shape
df.describe()
df.info()
```

```python
# Problem 2
# Fill missing values:
# - price -> mean price
# - quantity -> median quantity
df["price"] = df["price"].fillna(df["price"].mean())
df["quantity"] = df["quantity"].fillna(df["quantity"].median())

df["total_price"] = df["price"] * df["quantity"]
print(df)
```

```python
df.corr(numeric_only = True)
```

```python
# Bonus 4
# Find all customers whose total spending
# is above the average customer spending.
customer_spending = df.groupby("customer")["total_price"].sum()

customer_spending[
    customer_spending > customer_spending.mean()
]
```

```python
# Bonus 5
# Find the top category within each city
# based on total revenue.
df.groupby(["city", "category"])["total_price"].sum()
```
